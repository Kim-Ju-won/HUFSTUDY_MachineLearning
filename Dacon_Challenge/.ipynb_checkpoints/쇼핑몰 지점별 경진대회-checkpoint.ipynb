# 1. 데이터 분석

- 학습 데이터, 및 트레인 데이터를 불러와 학습을 합니다. 이 때 컬럼의 갯수와 이름, 타겟 데이터 등을 확인할 수 있어야 합니다. 추가로 시각화를 통해 데이터 형태를 파악하는 것도 좋습니다. 데이터를 분석한 뒤 모델링을 진행해주세요! 

## 1.1 패키지 다운로드 

- 필요한 패키지를 다운로드 합니다. 이 때 다운로드 받는 패키지 및 라이브러리는 상단에 모두 모아 두는게 좋습니다. 

````python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt # 시각화 => 그래프 (막대, 시계열)
import seaborn as sns # 시각화 => 자동으로 통계 그래프 (correlation 그래프, 히스토그램, heatmap)
import statsmodels.api as sm # 선형회귀 함수
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor # 다중공산성
````

## 1.2 데이터 로드 

- 학습 데이터와 타겟 데이터를 로드 
```bash
id : 샘플 아이디
Store : 쇼핑몰 지점
Date : 주 단위(Weekly) 날짜
Temperature : 해당 쇼핑몰 주변 기온
Fuel_Price : 해당 쇼핑몰 주변 연료 가격
Promotion 1~5 : 해당 쇼핑몰의 비식별화된 프로모션 정보
Unemployment : 해당 쇼핑몰 지역의 실업률
IsHoliday : 해당 기간의 공휴일 포함 여부
Weekly_Sales : 주간 매출액 (목표 예측값)
```

````python
train_data = pd.read_csv("./dataset/train.csv")
train_data
````

````python
test_data = pd.read_csv("./dataset/test.csv")
test_data
````

## 1.3 데이터 전처리

- 결측치를 제거하거나(0으로 채우거나), One-hot encoding 으로 범주형 데이터를 수치형으로 변환을 하고, 정규화를 진행합니다. 

````python
# 결측치 확인
train_data.isna().sum()
````

- 프로모션에 여러값들이 들어있고, 전체 데이터가 6000개정도 밖에 되지 않기 때문에 단순히 결측치를 삭제하는 건 좋은 방법이 아닌듯 합니다. 0으로 채워주겠습니다.
- 또한 Target Data인 Weekly Sales가 다 비워져있는 것도 알 수 있네요

````python
# 결측치 0으로 채우기
train_data=train_data.fillna(0)
train_data.isna().sum()
````

````python
print(train_data.columns)
print(train_data.dtypes)
print("행 열 :", train_data.shape)
````

- Bool 변수를 one-hot encoding을 통해서 0,1값으로 바꿔줍니다.

````python
train_data["IsHoliday"] = train_data["IsHoliday"].astype(int)
````

- pandas에 자동으로 histogram을 만들어 주는 함수인 .hist를 이용해서 histogram을 만들어줍니다. 

````python
train_data.hist(figsize=(30,20))
````

- 이제 데이터의 특성을 수치형으로 확인합니다. 

````python
train_data.describe()
````

- 정규화를 적용시켜 줍니다. 정규화는 Z- 정규화를 이용할 것인데요. 데이터 분포를 통해 확인할 수 있듯이, 많은 이상치가 존재합니다. min_max를 이용할 경우 이상치에 여전히 민감하기 때문에 Z-정규화를 이용해줍니다. 

````python
# id 는 제외하고 정규화 진행
norm = train_data.drop(['id','Date', 'Weekly_Sales'],axis=1)
# z-정규화( x-평균/표준편차)
train_data_normed = (norm- norm.mean())/norm.std()
train_data_normed
````

````python
train_data_normed.hist(figsize=(30,20))
````

- 시각화를 통해 데이터를 분석합니다. 

````python
temp = pd.merge(train_data_normed, train_data['Weekly_Sales'],
                left_index = True, right_index=True)
````

````python
temp
````

````python
# 선형성 확인
plt.figure(figsize=(16,16))
sns.heatmap(temp.corr(), linewidths=.5, cmap = 'Blues', annot=True)
````

# 2. 데이터 모델링

- 선형 회귀 모델 만들기 
- 종속 변수와 타겟 데이터 설정하기

## 2.1 데이터 학습

````python
y_target = train_data['Weekly_Sales']
x_data = train_data_normed.drop(['Unemployment','IsHoliday','Promotion4'],axis=1)

train_x, test_x, train_y, test_y = train_test_split(x_data, y_target, train_size=0.8, test_size=0.2,random_state = 2)
````

````python
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape)
````

````python
# 임의의 변수로 초기화 뒤 학습
test_x=sm.add_constant(test_x,has_constant='add')# 상수항추가
train_x=sm.add_constant(train_x,has_constant='add')# 상수항
model = sm.OLS(train_y, train_x)
fitted_model = model.fit()
````

````python
fitted_model.summary()
````

- 다중 공산성 체크

````python
# 다중 공산성 확인
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(
    x_data.values, i) for i in range(x_data.shape[1])]
vif["features"] = x_data.columns
vif
````

- 학습 데이터 정확도 시각화 

````python
plt.plot(np.array(fitted_model.predict(train_x)),label="pred")
plt.plot(np.array(train_y),label="true")
plt.legend()
plt.show()
````

- 테스트 데이터 정확도 시각화

````python
plt.plot(np.array(fitted_model.predict(test_x)),label="pred")
plt.plot(np.array(test_y),label="true")
plt.legend()
plt.show()
````

## 2.2 최종 Test Data 체크 및 Submission Data 생성

````python
#전처리
test_data["IsHoliday"] = test_data["IsHoliday"].astype(int)
test_data=test_data.fillna(0)
# id 는 제외하고 정규화 진행
normed = test_data.drop(['id','Date'],axis=1)
# z-정규화
test_data_normed = (normed- normed.mean())/normed.std()
test_data_normed=test_data_normed.fillna(0)
test_data_normed
````

````python
test_data_normed = test_data_normed.drop(['Unemployment','IsHoliday','Promotion4'],axis=1)
````

````python
predict_variable=sm.add_constant(test_data_normed,has_constant='add')
````

````python
predict = fitted_model.predict(predict_variable)
````

# 3. 제출 파일 만들기

````python
sample_submission = pd.read_csv('./dataset/sample_submission.csv')
sample_submission['Weekly_Sales'] = predict
sample_submission.to_csv('submission.csv',index = False)
sample_submission.head()
````
